# Pre-Processing the Absenteeism Dataset

### import the relevant libraries

```python
import pandas as pd
```

### Load the "Absenteeism Dataset"

```python
raw_csv_data = pd.read_csv('Absenteeism-data.csv')
```

```python
df = raw_csv_data.copy()
df
```

### cheking for missing values

```python
df.info()
```

## Drop 'ID':

```python
# since ID does not have any predictive power in our case we'll drop it
df = df.drop(['ID'], axis = 1)
df.head()
```

## 'Reason for Absence':

```python
df['Reason for Absence'].min()
```

```python
df['Reason for Absence'].max()
```

```python
df['Reason for Absence'].unique()
```

```python
len(df['Reason for Absence'].unique())
```

there is 28 distinct element and the length is 28 so all values from 1 to 28 are included

### creating dummies for 'Reason for absence'

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns = reason_columns.astype(int)
reason_columns
```

### cheking for missing values

```python
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns.head()
```

```python
reason_columns['check'].unique()
```

 then no missing values

```python
reason_columns.head()
```

```python
reason_columns = reason_columns.drop([0], axis = 1)
reason_columns
```

## Group the Reasons for Absence:

```python
# Reason from 1 to 14 is for Several Diseases and will be marked as reason_type_1
# Reason from 15 to 17 is for pregnancy and making birth and will be marked as reason_type_2
# Reason from 18 to 21 is for poisining and will be marked as reason_type_3
# Reason from 22 to 28 is for Light reasons(blood donation, dentist consultation...) and will be marked as reason_type_4

reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
df = df.drop(['Reason for Absence'], axis = 1)


df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
df
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']

df.columns = column_names
df.head()
```

## Reorder Columns

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']

df = df[column_names_reordered]
df.head()
```

## Checkpoint

```python
df_reason_mod = df.copy()
df_reason_mod
```

## 'Date':

```python
type(df_reason_mod['Date'][0])
```

```python
# we want to convert it from String to Timestamp to extract the month value and the weekday from it
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
df_reason_mod['Date'][0]
```

## Extract the Month Value:

```python
df_reason_mod.shape
```

```python
list_months = []

for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)

list_months
```

```python
len(list_months)
```

```python
pd.unique(list_months)
```

```python
df_reason_mod['Month Value'] = list_months
df_reason_mod.head()
```

## Extract the Day of the Week:

```python
def date_to_weekday(date_value):
    return date_value.weekday()

df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head()
```

### drop the 'Date' column 

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis = 1)
df_reason_mod.head()
```

### Reorder the columns 

```python
column_names_upd = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']

df_reason_mod = df_reason_mod[column_names_upd]
df_reason_mod.head()
```

### Checkpoint

```python
df_reason_date_mod = df_reason_mod.copy()
df_reason_date_mod
```
